##Question -01

```python
import pandas as pd
import matplotlib.pyplot as plt

# ----- Task 3: Population -----
# Load dataset
pop_df = pd.read_csv('/content/population.csv')  # must have 'Country', 'Population'

# Sort by population descending and take top 5
top5 = pop_df.sort_values('Population', ascending=False).head(5)

# Plot
plt.figure(figsize=(8,5))
bars = plt.bar(top5['Country'], top5['Population']/1_000_000_000)  # convert to billions
plt.title('Top 5 Countries by Population')
plt.ylabel('Population (Billions)')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

largest_country = top5.iloc[0]['Country']
print(f"The country with the largest population is: {largest_country}")
```

##Question-02

```python
# ----- Task 4: Movies by Genre -----
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
movies_df = pd.read_csv('/content/movies_genre.csv')  # must have 'Genre', 'MovieCount'

# Sort by movie count
movies_df = movies_df.sort_values('MovieCount', ascending=False)

# Plot
plt.figure(figsize=(10,6))
plt.bar(movies_df['Genre'], movies_df['MovieCount'])
plt.title('Number of Movies Produced by Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

most_popular_genre = movies_df.iloc[0]['Genre']
print(f"The genre with the most movies is: {most_popular_genre}")
```
